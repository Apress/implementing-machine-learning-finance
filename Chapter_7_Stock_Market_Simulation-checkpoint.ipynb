# Chapter 7: Stock Market Simulation

In this chapter, we implement Monte Carlo simulation to simulate considerable changes in the market without exposing ourselves to risk. Simulating the market helps identify patterns in preceding occurrences and forecast future prices with reasonable certainty. We assume the stock exchange market exists in the actual world. If we can reconstruct the actual world, then we can understand preceding market behavior and predict future market behavior. In the preceding chapters, we sufficiently covered models for sequential pattern identification and forecasting. We use the panda_montecarlo framework to perform simulation. # Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set("talk","ticks",font_scale=1,font="Calibri")
from pylab import rcParams
plt.rcParams["figure.dpi"] = 300
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import datetime as dt
import scipy
import pandas_montecarlo
from pandas_datareader import data
from scipy.stats import norm
```

```python
tickers = ['AMZN','AAPL','WBA',
           'NOC','BA','LMT']
start_date = '2010-01-01'
end_date = '2020-11-01'
df = data.get_data_yahoo(tickers, start_date, end_date)[['Adj Close']]
df.head()
```

# Data Preprocessing

```python
weights = np.array([.25, .3, .15, .10, .24, .7])
initial_investment = 5000000
```

```python
returns = df.pct_change()
returns.tail()
```

```python
cov_matrix = returns.cov()
cov_matrix
```

# 10-Days VAR

```python
conf_level1 = 0.05
avg_rets = returns.mean()
port_mean = avg_rets.dot(weights)
port_stdev = np.sqrt(weights.T.dot(cov_matrix).dot(weights))
mean_investment = (1+port_mean) * initial_investment
stdev_investment = initial_investment * port_stdev
cutoff1 = norm.ppf(conf_level1, mean_investment, stdev_investment)
var_1d1 = initial_investment - cutoff1
var_1d1
```

```python
var_array = []
num_days = int(10)
for x in range(1, num_days+1):    
    var_array.append(np.round(var_1d1 * np.sqrt(x),2))
    print(str(x) + " day VaR @ 95% confidence: " + str(np.round(var_1d1 * np.sqrt(x),2)))
```

## Maximum Porfolio Loss

```python
plt.plot(var_array, color="navy")
plt.xlabel("No. of Days")
plt.ylabel("MaX Portfolio Loss (USD)")
plt.show()
```

## Develop Monte Carlo Simmulation

```python
start_date = '2010-11-01'
end_date = '2020-11-01'
ticker = 'NOC'
df = data.get_data_yahoo(ticker, start_date, end_date)
df['return'] = df['Adj Close'].pct_change().fillna(0)
df.head()
```

```python
mc = df['return'].montecarlo(sims=10, bust=-0.1, goal=1)
mc.plot(title="")
```
