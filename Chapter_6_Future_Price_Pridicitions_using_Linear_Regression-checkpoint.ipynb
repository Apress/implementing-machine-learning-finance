# Chapter 6: Future Price Prediction using Linear Regression

This chapter introduces the parametric method, equally recognized as the linear regression method. We use this method to determine the nature of the relationship between an independent variable (continuous or categorical) and a dependent variable (always continuous). 

In this chapter, we predict the Closing Price of Gold by the highest and lowest price change and returns.

# Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set("talk","ticks",font_scale=1,font="Calibri")
from pylab import rcParams
plt.rcParams["figure.dpi"] = 300
import warnings
warnings.filterwarnings("ignore")
import statsmodels.api as sm
from pandas_datareader import data
start_date = '2010-11-01'
end_date = '2020-11-01'
ticker = 'GC=F'
df = data.get_data_yahoo(ticker, start_date, end_date)
df_orig=df
#calculate highest and lowest price change
df['HL_PCT']=(df['High']-df['Low'])/df['Adj Close'] *100.0
#calculate return of stock price
df['PCT_change']= (df['Adj Close']-df['Open'])/df['Open'] *100.0
df = df[['Adj Close','HL_PCT','PCT_change','Volume']]

date = df.index
df.head()
```

# Data Preprocessing

```python
from sklearn import preprocessing
x=df[['Adj Close','HL_PCT','Volume']]
x=preprocessing.scale(x)
y=df['Adj Close']
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x , y, test_size=0.1,shuffle= False)
```

# Develop the OLS Model

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression(n_jobs=-1)
lm.fit(x_train, y_train)
```

## Predictions

```python
y_pred = lm.predict(x_test)
pd.DataFrame(y_pred,columns=["Forecast"])
```

```python
num_samples = df.shape[0]
#add Forecase column to dataframe
df['Forecast'] = np.nan
df['Forecast'][int(0.9*num_samples):num_samples]=y_pred
df['Adj Close'].plot(color="navy")
df['Forecast'].plot(color="red")
plt.legend(loc="best")
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

# Model Evaluation

```python
from sklearn import metrics
MAE = metrics.mean_absolute_error(y_test,y_pred)
MSE = metrics.mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
R2 = metrics.r2_score(y_test,y_pred)
EV = metrics.explained_variance_score(y_test,y_pred)
MGD = metrics.mean_gamma_deviance(y_test,y_pred)
MPD = metrics.mean_poisson_deviance(y_test,y_pred)
lmmodelevaluation = [[MAE,MSE,RMSE,R2,EV,MGD,MPD]]
lmmodelevaluationdata = pd.DataFrame(lmmodelevaluation,
                                     index = ["Values"],
                                     columns = ["MAE",
                                                "MSE",
                                                "RMSE",
                                                "R2",
                                                "Explained variance score",
                                                "Mean gamma deviance",
                                                "Mean Poisson deviance"]).transpose()
lmmodelevaluationdata
```
