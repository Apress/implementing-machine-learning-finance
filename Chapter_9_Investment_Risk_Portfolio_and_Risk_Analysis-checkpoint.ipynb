# Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set("talk","ticks",font_scale=1,font="Calibri")
from pylab import rcParams
plt.rcParams["figure.dpi"] = 300
import warnings
warnings.filterwarnings("ignore")
```

```python
from pandas_datareader import data
import pyfolio as pf
ticker = 'AMZN'
start_day = '2010-10-01'
end_day = '2020-10-01'
amzn = data.get_data_yahoo(ticker, start_day, end_day)
spy = data.get_data_yahoo('SPY', start_day,end_day)
```

# Data Preprocessing

```python
amzn = amzn["Adj Close"].pct_change()[1:]
spy = spy["Adj Close"].pct_change()[1:]
```

# Performance Statistics

```python
pf.show_perf_stats(amzn, spy)
```

# Drawdown Underwater

```python
pf.plot_drawdown_underwater(amzn)
plt.title("")
plt.show()
```

## Annual Returns

```python
pf.plot_annual_returns(amzn)
plt.title("")
plt.show()
```

## Rolling Returns

```python
pf.plot_rolling_returns(amzn)
plt.title("")
plt.show()
```
